                <div class="subsection">
                    <h3 class="subsection-title">Our Communication Framework</h3>
                    <p class="content-text">
                        Effective communication is critical for team success. We will use a combination of synchronous and asynchronous communication channels to ensure everyone stays informed and engaged throughout the project lifecycle.
                    </p>
                    
                    <p class="content-text"><strong>Communication Channels:</strong></p>
                    <ul class="benefits-list">
                        <li><strong>Daily Slack Channel:</strong> Quick updates, questions, and informal discussion happen in our dedicated team Slack channel</li>
                        <li><strong>GitHub Issues & Pull Requests:</strong> All code-related discussion happens directly in GitHub where context is preserved</li>
                        <li><strong>Weekly Video Meetings:</strong> We meet via Zoom every Monday and Thursday at 4 PM for sprint planning and retrospectives</li>
                        <li><strong>Shared Google Docs:</strong> Planning documents, meeting notes, and design decisions are documented in shared Google Docs</li>
                        <li><strong>Trello Board:</strong> Our project management board shows all tasks, assignments, and progress in real-time</li>
                    </ul>
                </div>
                
                <div class="subsection">
                    <h3 class="subsection-title">Scrum Agile Methodology</h3>
                    <p class="content-text">
                        We will implement Scrum Agile practices to organize our work into manageable sprints, maintain momentum, and ensure continuous improvement. Our sprints will be one week long, allowing for rapid iteration and frequent delivery of working features.
                    </p>
                    
                    <div class="scrum-methodology">
                        <div class="scrum-ceremony">
                            <h4>üìã Sprint Planning (Monday, 4:00 PM)</h4>
                            <p>At the start of each week, we meet to review the backlog, select tasks for the upcoming sprint, estimate effort, and assign responsibilities. Each team member commits to specific deliverables.</p>
                        </div>
                        
                        <div class="scrum-ceremony">
                            <h4>‚è∞ Daily Standup (Async via Slack, 9:00 AM)</h4>
                            <p>Every morning, each team member posts a brief update answering three questions: What did I accomplish yesterday? What will I work on today? Are there any blockers? This keeps everyone aligned without requiring real-time meetings.</p>
                        </div>
                        
                        <div class="scrum-ceremony">
                            <h4>üîç Sprint Review (Thursday, 4:00 PM)</h4>
                            <p>Mid-week, we demonstrate completed features to each other, gather feedback, and discuss any adjustments needed. This ensures we catch issues early and celebrate progress.</p>
                        </div>
                        
                        <div class="scrum-ceremony">
                            <h4>üîÑ Sprint Retrospective (End of Thursday Meeting)</h4>
                            <p>After reviewing our work, we reflect on our process: What went well? What could be improved? What will we change next sprint? This continuous improvement mindset helps us work more effectively over time.</p>
                        </div>
                        
                        <div class="scrum-ceremony">
                            <h4>üìä Backlog Grooming (As Needed)</h4>
                            <p>Throughout the week, we refine upcoming tasks, break down large features into smaller tickets, and update priorities based on new information or feedback.</p>
                        </div>
                    </div>
